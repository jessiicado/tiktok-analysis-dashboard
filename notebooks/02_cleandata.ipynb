```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("../data/raw/tiktok_data.csv") #read the dataset
print(f"Raw shape: {df.shape}")
```

```python
# drop any rows with missing caption
df = df.dropna(subset=["text_part"])
print(f"After dropping null captions: {df.shape}")
```

```python
#fix data types
df["views"] = pd.to_numeric(df["views"], errors="coerce")
df["likes"] = pd.to_numeric(df["likes"], errors="coerce")
df["author_followers"] = pd.to_numeric(df["author_followers"], errors="coerce")
df["author_likes"] = pd.to_numeric(df["author_likes"], errors="coerce")
df["duration_seconds"] = pd.to_numeric(df["duration_seconds"], errors="coerce")
df["human_time"] = pd.to_datetime(df["human_time"], errors="coerce")
print("Data types fixed")
df.dtypes
```

```python
#drop rows with null cols
df = df.dropna(subset=["views", "likes", "author_followers", "duration_seconds"])
print(f"After dropping null key columns: {df.shape}")
```

```python
#remove all duplicates
df = df.drop_duplicates(subset=["id_video"])
print(f"After removing duplicates: {df.shape}")
```

```python
#engagement rate col
df["engagement_rate"] = df["likes"] / df["views"].replace(0, np.nan)
print("Engagement rate column added")
df[["views", "likes", "engagement_rate"]].head()
```

```python
#date parts for time analysis
df["post_date"] = df["human_time"].dt.date
df["post_hour"] = df["human_time"].dt.hour
df["post_month"] = df["human_time"].dt.month
print("Date columns added")
```

```python
# cleaning caption
df["text_part"] = df["text_part"].astype(str).str.strip()
print("Caption text cleaned")
```

```python
print(f"Final shape: {df.shape}")
print(f"\nNull values remaining:")
print(df.isnull().sum())
```

```python
# save to processed
df.to_csv("../data/processed/tiktok_clean.csv", index=False)
print("Saved to data/processed/tiktok_clean.csv")
```
